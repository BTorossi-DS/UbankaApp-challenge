```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns

plt.style.use('seaborn-dark')
```

```python
catalog_data = pd.read_csv(r'C:\Users\Win10\data\catalog.csv')

projects_data = pd.read_csv(r'C:\Users\Win10\data\test_projects.csv')

rules_data = pd.read_csv(r'C:\Users\Win10\data\test_rules.csv')

transactions_data = pd.read_csv(r'C:\Users\Win10\data\test_transactions.csv')
catalog_data
```

```python
# df['hasimage'] = np.where(df['photos']!= '[]', True, False)

catalog_data['project_category_id'] = np.where(catalog_data['type'] == 'project_categories', catalog_data['id'], '-')
catalog_data['rule_type_id'] = np.where(catalog_data['type'] == 'rule_types', catalog_data['id'], False)
catalog_data['merchant_id'] = np.where(catalog_data['type'] == 'merchants', catalog_data['id'], False)

catalog_data = catalog_data.rename(columns={'name': 'category_name'})
catalog_data
```

```python
sns.boxplot(x=projects_data['total'])
```

<b><u> Removing outliers

```python
outliers = projects_data[projects_data.total > 100000].count()

projects_data_clean = projects_data[projects_data.total < 100000]
projects_data_clean
```

```python
projects_objetives_amounts = sns.boxplot(x=projects_data_clean['total']).set_title('Projects objetives amounts')
projects_objetives_amounts
```

```python
merged_projects= pd.merge(projects_data_clean,catalog_data, on='project_category_id')
merged_projects = merged_projects.drop(columns= ['id','rule_type_id','merchant_id', 'type'])
merged_projects
```

```python
projects_by_category = merged_projects[['project_id', 'category_name']].groupby(['category_name'], 
                                                                                as_index= False).count()

projects_by_category = projects_by_category.rename(columns={'category_name': 'Project_category',
                                                            'project_id': 'Total_projects'})
projects_by_category = pd.DataFrame(projects_by_category)
projects_by_category
```

```python
import matplotlib.pyplot as plt
plt.bar(projects_by_category['Project_category'], projects_by_category['Total_projects'])
plt.title('Numbers of projects by category')
```

```python

merged_rules= pd.merge(rules_data,catalog_data, on='rule_type_id')
merged_data = pd.merge(merged_rules, merged_projects, on='project_id')

merged_rules = merged_rules.drop(columns= ['id','project_category_id','merchant_id', 'type'])
merged_data = merged_data.drop(columns= ['id','project_category_id_y',
                                         'merchant_id', 'type',
                                         'project_category_id_x', 'rule_type_id'
                                        ])

merged_data = merged_data.rename(columns={'category_name_x': 'Rule', 'category_name_y': 'Project Category'})

merged_rules.groupby(['category_name'], as_index= False).count()
merged_data
```

```python
merged_transactions = pd.merge(transactions_data,merged_data, on='user_id')
merged_transactions
```

```python
merged_data.groupby(['Project Category'], as_index= False).count()
```

```python
avg_projects_obj = merged_projects[['total', 'category_name']].groupby(['category_name'], as_index= False).mean().round()
avg_projects_obj = avg_projects_obj.rename(columns={'category_name': 'Project_category',
                                                    'total': 'Average Objetive'})
avg_projects_obj
```

```python
plt.bar(avg_projects_obj['Project_category'], avg_projects_obj['Average Objetive'])
plt.title('Average objetive amount by category')
```
